```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns

# Load dataset
data = pd.read_csv('my_movies.csv')
```

```python
# Convert the dataset to a transaction format
transactions = data.drop(['V1', 'V2', 'V3', 'V4', 'V5'], axis=1)

transactions = transactions.astype(bool)

transactions
```

```python
# Function to create a network plot
def plot_network(rules, figsize=(8, 6)):
    G = nx.DiGraph()

    for i in range(len(rules)):
        G.add_node(rules.iloc[i]['antecedents'], color='skyblue', node_size=300)
        G.add_node(rules.iloc[i]['consequents'], color='salmon', node_size=300)
        G.add_edge(rules.iloc[i]['antecedents'], rules.iloc[i]['consequents'], color='gray', alpha=0.5)

    pos = nx.spring_layout(G)
    plt.figure(figsize=figsize)
    nx.draw(G, pos, with_labels=True, font_size=6, node_color=[G.nodes[n]['color'] for n in G.nodes], edge_color='gray', font_color='black', node_size=[G.nodes[n]['node_size'] for n in G.nodes])
    plt.title('Association Rules Network Plot')
    plt.show()

# Apply Apriori algorithm with different Support, Confidence and Minimum length values
support_values = [0.2, 0.3, 0.4]
confidence_values = [0.5, 0.6, 0.7]
min_length_values = [1, 2]

# Main loop for support and confidence values
for support in support_values:
    for confidence in confidence_values:
        
        # Apply Apriori algorithm without specifying min_len initially
        frequent_itemsets = apriori(transactions, min_support=support, use_colnames=True)
        
        # Inner loop for different minimum length values
        for min_length in min_length_values:
            
            # Filter out itemsets with length less than min_length
            frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
            frequent_itemsets_filtered = frequent_itemsets[frequent_itemsets['length'] >= min_length]
            frequent_itemsets_filtered = frequent_itemsets_filtered.drop('length', axis=1)
            
            # Check if the filtered frequent itemsets DataFrame is not empty
            if not frequent_itemsets_filtered.empty:

                print(f"\n Frequent Itemsets with min_length={min_length}, support={support}, confidence={confidence}:")
                print(frequent_itemsets_filtered)
                print("\n")
                
                try:
                    # Generate association rules based on the filtered frequent itemsets
                    rules = association_rules(frequent_itemsets_filtered, metric="confidence", min_threshold=confidence)

                    print(f"Association Rules with min_length={min_length}, support={support}, confidence={confidence}:")
                    print(rules)
                    print("\n")
                    
                    # Visualize the rules using the network plot
                    print(f"Min length={min_length}, Support={support}, Confidence={confidence}:")
                    plot_network(rules)
                    print("\n")
                    
                    # Plotting support vs confidence
                    print(f"Min length={min_length}, Support={support}, Confidence={confidence}:")
                    plt.figure(figsize=(8, 5))
                    plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
                    plt.xlabel('Support')
                    plt.ylabel('Confidence')
                    plt.title('Support vs Confidence')
                    plt.show()
                    
                    # Histogram
                    plt.figure(figsize=(8, 5))
                    print(f"Min length={min_length}, Support={support}, Confidence={confidence}:")
                    plt.hist(rules['lift'], bins=20, color='green', alpha=0.7)
                    plt.title('Histogram: Lift')
                    plt.xlabel('Lift')
                    plt.ylabel('Frequency')
                    plt.show()
                    
                    # Heatmap for Support, Confidence, and Lift
                    print(f"Min length={min_length}, Support={support}, Confidence={confidence}:")
                    heatmap_data = rules[['support', 'confidence', 'lift']]
                    plt.figure(figsize=(8, 5))
                    sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt=".3f")
                    plt.title('Heatmap: Support, Confidence, and Lift')
                    plt.show()

                except KeyError:
                    # If KeyError occurs, try using support_only=True
                    rules = association_rules(frequent_itemsets_filtered, metric="confidence", min_threshold=confidence, support_only=True)

                    if not rules.empty:
                        print(f"Association Rules with min_length={min_length}, support={support}, confidence={confidence} (support_only=True):")
                        print(rules)
                    else:
                        print(f"Rule is empty for min_length={min_length}, support={support}, confidence={confidence}")
                    
                    print("\n")
            else:
                print(f"\n No frequent itemsets with min_length={min_length}, support={support}, confidence={confidence}\n")
```
