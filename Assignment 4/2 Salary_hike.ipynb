```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Read the data

data = pd.read_csv("Salary_Data.csv")

print(data.head())
```

```python
# Check the info of the DataFrame
data.info()
```

```python
# Perform Exploratory Data Analysis (EDA)
data.describe()
```

```python
sns.distplot(data['YearsExperience'])
```

```python
sns.distplot(data['Salary'])
```

```python
# Histogram plot for YearsExperience
sns.histplot(data['YearsExperience'], bins=10, kde=True, color='lightblue')
plt.title('Histogram of Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.show()

# Histogram plot for Salary
sns.histplot(data['Salary'], bins=10, kde=True, color='lightgreen')
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
data.corr()
```

```python
# Scatter plot to visualize the relationship
plt.scatter(data['YearsExperience'], data['Salary'])
plt.title('Salary vs YearsExperience')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# Model Building

# Fit a simple linear regression model
model = LinearRegression()
X = data[['YearsExperience']]
y = data['Salary']
model.fit(X, y)
```

```python
# Display the coefficients and intercept
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
```

```python
# Make predictions
test_data_for_prediction = pd.DataFrame({'YearsExperience': [3, 6]})
predictions = model.predict(test_data_for_prediction)
print(predictions)
```
