```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Read the data

data = pd.read_csv("delivery_time.csv")

print(data.head())
```

```python
# Check the info of the DataFrame
data.info()
```

```python
# Perform Exploratory Data Analysis (EDA)
data.describe()
```

```python
sns.distplot(data['Sorting Time'])
```

```python
sns.distplot(data['Delivery Time'])
```

```python
# Histogram

# Univariate plot for Delivery Time
sns.histplot(data['Delivery Time'], bins=10, kde=True, color='lightblue')
plt.title('Histogram of Delivery Time')
plt.xlabel('Delivery Time')
plt.ylabel('Frequency')
plt.show()

# Univariate plot for Sorting Time
sns.histplot(data['Sorting Time'], bins=10, kde=True, color='lightgreen')
plt.title('Histogram of Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Frequency')
plt.show()
```

```python
data.corr()
```

```python
# Scatter plot to visualize the relationship
plt.scatter(data['Sorting Time'], data['Delivery Time'])
plt.title('Delivery Time vs Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

```python
# Model Building

# Fit a simple linear regression model
model = LinearRegression()
X = data[['Sorting Time']]
y = data['Delivery Time']
model.fit(X, y)
```

```python
# Display the coefficients and intercept
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
```

```python
# Make predictions
test_data_for_prediction = pd.DataFrame({'Sorting Time': [5, 8]})
predictions = model.predict(test_data_for_prediction)
print(predictions)
```
